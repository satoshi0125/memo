```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp = np.array([36.7,36.9,37.1,37.2,37.6,None,38.1,37.7,37.7,37.6,37.9,37.3,38.2,37.5,36.6,36.9,36.7,36.8])
time = np.arange(0,len(temp)*3,3)
```

```python
plt.figure(figsize=(5,3),dpi=100)
plt.plot(time,temp,label='measured value')
plt.scatter(time,temp)
plt.xlabel('elapsed time [hour]',fontsize=15)
plt.ylabel('body temperature [℃]',fontsize=15)
plt.xlim(0,len(temp)*3)
plt.ylim(36,39)
plt.plot(time,list([36.6])*len(time),color='green',label='normal value')
plt.hlines(np.arange(36, 50, 0.5),0,50,linestyle='dashed',color='silver')
plt.legend()
plt.show()
```

## 所感
- 症状：発熱、寒気、倦怠感、接種した腕の痛み
- 症状期間：約日
- 対応策：寝る、汗拭く、水分接種、冷えピタ、解熱剤
- できたこと：家事（洗濯、料理、掃除）
- できなかったこと：運動、飲酒
